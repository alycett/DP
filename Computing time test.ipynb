```python
import cantera as ct
import numpy as np
import time
```

```python
# Methane
gas01 = ct.Solution('gri30.yaml')
# Octane
gas02 = ct.Solution('Jerzembeck.yaml')
# Ethanol and Octane
#gas03 = ct.Solution('prf-curran.yaml')
# Ethanol and Octane
gas03 = ct.Solution('octane-mehl.yaml')
```

```python
# Methane
start = time.time()
gas01.TPX = 300.0, ct.one_atm, 'CH4:1,O2:2,N2:7.52'
gas01.equilibrate('HP')
stop = time.time()
elapsed_time = stop - start

print('The equilibrate function took', elapsed_time, 'seconds to compute')
print('The equilibirum temperature is', gas01.T, '\n')
```

```python
# Iso-octane, Jerzembeck
start = time.time()
gas02.TPX = 300.0, ct.one_atm, 'IXC8H18:1,O2:12.5,N2:47'
gas02.equilibrate('HP')
stop = time.time()
elapsed_time = stop - start

print('The equilibrate function took', elapsed_time, 'seconds to compute')
print('The equilibirum temperature is', gas02.T, '\n')
```
